```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
lines = list()
with open(".\\output.txt", "r") as file:
    lines = file.readlines()
```

```python
block_size = []
time_omac = []
time_hmac = []

for line in lines:
    buf = line.split(' ')
    block_size.append(int(buf[0]))
    time_omac.append(float(buf[1]))
    time_hmac.append(float(buf[2]))
```

```python
plt.figure(figsize=(10, 7))
plt.plot(block_size, time_omac, 'go-', label="OMAC")
plt.plot(block_size, time_hmac, 'bo-', label="HMAC")
plt.legend()
plt.xscale("log")
plt.title('Среднее время вычисления OMAC и HMAC, логарифмическая шкала')
plt.xlabel('Размер входного сообщения, байты')
plt.ylabel('Время вычисления, миллисекунды')
plt.savefig(".\\plot")
```
