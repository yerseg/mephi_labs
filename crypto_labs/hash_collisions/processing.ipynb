```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
lines = list()
with open(".\\out_files\\birth_mem_&_time.txt", "r") as file:
    lines = file.readlines()
```

```python
sha_size = []
time = []
mem = []
for line in lines:
    buf = line.split(' ')
    sha_size.append(int(buf[0]))
    time.append(float(buf[1]))
    mem.append(float(buf[2]))
```

```python
plt.figure(figsize=(10, 7))
plt.plot(sha_size, time, 'bo')
plt.title('Среднее время нахождения коллизии на основе парадокса ДР')
plt.xlabel('Размер выхода хэш-функции, биты')
plt.ylabel('Время вычисления, миллисекунды')
plt.savefig(".\\birth_time_plot")
```

```python
plt.figure(figsize=(10, 7))
plt.plot(sha_size, mem, 'go')
plt.title('Средняя оценка затрат памяти на нахождение коллизии')
plt.xlabel('Размер выхода хэш-функции, биты')
plt.ylabel('Выделенная память, байты')
plt.savefig(".\\birth_mem_plot")
```
